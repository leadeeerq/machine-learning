```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

#Generowanie danych

```python
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1000, centers=None, cluster_std=1.0, center_box=(-4.0, 4.0), random_state=42)[0]
df = pd.DataFrame(data, columns=['x1', 'x2'])
df.head()
```

#Wizualizacja danych

```python
# px.scatter(df, 'x1', 'x2', width=950, height=500, title='Klasteryzacja - algorytm K-średnich')
plt.scatter(df['x1'], df['x2'])
plt.rcParams["figure.figsize"] = (20,3)
plt.title('Klasteryzacja - algorytm K-średnich')
```

#Algorytm K-średnich

```python
from sklearn.cluster import KMeans

n_clusters = 6

kmeans = KMeans(n_clusters= n_clusters)
kmeans.fit(data)
```

```python
y_kmeans = kmeans.predict(data)
y_kmeans[:10]
```

```python
df['y_kmeans'] = y_kmeans
df.head()
```

#Wizualizacja

```python
df['y_kmeans'][2]
```

```python
px.scatter(df, 'x1', 'x2', 'y_kmeans', width=950, height=500, title=f'Algorytm K-średnich - {n_clusters} klastry')
```
